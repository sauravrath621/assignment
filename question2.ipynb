```python
import cv2
import matplotlib.pyplot as plt
img1=cv2.imread('tejas 2.jpg')
plt.imshow(img1)
plt.show()
```

```python
(h,w)=img1.shape[:2]
(q_h,q_w)=(int(h/4),int(w/4))
```

```python
import numpy as np
m=np.float32([[1,0,q_w],[0,1,q_h]])
print(m)
```

```python
translated = cv2.warpAffine(img1,m,(w,h))
plt.imshow(translated)
plt.show()
```

```python
import numpy as np
m=np.float32([[1,0,q_w],[0,1,0]])
print(m)
```

```python
translated = cv2.warpAffine(img1,m,(w,h))
plt.imshow(translated)
plt.show()
```

```python
import numpy as np
m=np.float32([[1,0,0],[0,1,q_h]])
print(m)
translated = cv2.warpAffine(img1,m,(w,h))
plt.imshow(translated)
plt.show()
```
