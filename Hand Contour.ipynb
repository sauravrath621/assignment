```python
import cv2
import matplotlib.pyplot as plt
```

```python
img2=cv2.imread('hand.jpeg')
plt.imshow(img2)
plt.show()
```

```python
gaussian = cv2.GaussianBlur(img2,(3,3),0)
edge = cv2.Canny(gaussian,20,40)
plt.imshow(edge)
plt.show()
```

```python
contours, hierarchy = cv2.findContours(edge, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
print('Numbers of contours are : ', len(contours))
cv2.drawContours(img2, contours, -1, (255, 0, 0), 3)
plt.imshow(img2)
plt.show()
```
